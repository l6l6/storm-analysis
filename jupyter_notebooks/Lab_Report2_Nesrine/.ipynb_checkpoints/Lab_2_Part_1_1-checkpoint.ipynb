### Create an XML file for analysis

````python
maximum_number_of_frames = -1 # Read all frames

# background params
background_sigma = 11.0
foreground_sigma = 1.0
drift_correction = 0

pixel_size = 100.0
threshold = 6
sigma = 258.21/(100.0 * 2.355)
iterations = 20
roi_size= 11

# Camera params
camera_gain = 1.0
camera_offset = 100.0

# Peak params
find_max_radius = 5
````

### b. 2D Asymmetric Gaussian function


````python
from scipy.optimize import curve_fit
import numpy as np

def gaussian2D_asym(position, amplitude, x_center, y_center, sigma_x, sigma_y, theta):
    x,y = position
    a = (np.cos(theta)**2)/(2*sigma_x**2) + (np.sin(theta)**2)/(2*sigma_y**2)
    b = -(np.sin(2*theta))/(4*sigma_x**2) + (np.sin(2*theta))/(4*sigma_y**2)
    c = (np.sin(theta)**2)/(2*sigma_x**2) + (np.cos(theta)**2)/(2*sigma_y**2)
    g = amplitude*np.exp( - (a*((x-x_center)**2) + 2*b*(x-x_center)*(y-y_center)
                             + c*((y-y_center)**2)))
    return g.ravel()

def fit_2d_asym_gauss(image, params):
    """
    Fits a 2D asymmetric gaussian
    :param image: input image
    :param params: array of input parameters; [x_0, y_0, sigma_x, sigma_y, amp, phase]
    :return: data fitted
    """
    index_x, index_y = image.shape
    xmin, xmax = 0, 1
    ymin, ymax = 0, 1
    x = np.linspace(xmin, xmax, index_x)
    y = np.linspace(ymin, ymax, index_y)
    x, y = np.meshgrid(x, y)

    popt, pcov = curve_fit(gaussian2D_asym, (x, y), image.ravel(), params)

    data_fitted = gaussian2D_asym((x, y), *popt)
    return popt, data_fitted.reshape(index_x, index_y)
````

### c. and d. Analysis of a single frame

````python
import storm_analysis.sa_library.datareader as datareader
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import storm_analysis.sa_library.sa_h5py as saH5Py
import matplotlib.pyplot as pyplot
import matplotlib.patches as patches
import numpy

def overlay_image_loc(orig_name, locs_name, frame_number, margins=None, sx = 9, sy = 9):
    """
    Create an image of a frame with the localizations overlaid.

    orig_name - The name of the original file.
    locs_name - The name of the localizations HDF5 file.
    frame_number - Which frame to examine.
    sx - Figure x size in inches.
    sy - Figure y size in inches.
    """
    frame = datareader.inferReader(orig_name).loadAFrame(frame_number).astype(numpy.float64)
    with saH5Py.SAH5Py(locs_name) as h5:
        locs = h5.getLocalizationsInFrame(frame_number)

    frame = frame - numpy.min(frame)
    frame = frame/numpy.max(frame)

    fig = pyplot.figure(figsize = (sx, sy))
    ax = fig.add_subplot(1,1,1)
    fig.subplots_adjust(left=0.2, bottom=0.2)
    for i in range(locs["x"].size):
        width = 10
        height = 10
        if "xsigma" in locs:
            width = height = 5.0*locs["xsigma"][i]
        if "ysigma" in locs:
            height = 5.0*locs["ysigma"][i]
        ellipse = patches.Ellipse((locs["x"][i], locs["y"][i]), width, height, facecolor='none', edgecolor='r', linewidth = 1)
        ax.add_artist(ellipse)

    ax.set_title("Overlay Image on frame {}".format(frame_number))
    ax.imshow(frame, interpolation = 'nearest', cmap = "gray")
    ax.scatter(locs["x"], locs["y"], cmap="gray")

    if margins:
        axins = zoomed_inset_axes(ax, 2, loc='lower left', bbox_to_anchor=(0,0), borderpad=3)
        mark_inset(ax, axins, loc1=2, loc2=4)
        axins.set_xlim(margins, margins+20)
        axins.set_ylim(margins, margins+20)
        axins.imshow(frame, cmap = "gray")
        axins.scatter(locs["x"], locs["y"], cmap="gray")
        for i in range(locs["x"].size):
            width = 10
            height = 10
            if "xsigma" in locs:
                width = height = 5.0*locs["xsigma"][i]
            if "ysigma" in locs:
                height = 5.0*locs["ysigma"][i]
            ellipse = patches.Ellipse((locs["x"][i], locs["y"][i]), width, height, facecolor='none', edgecolor='r', linewidth = 1)
            axins.add_artist(ellipse)
    pyplot.show()
````

````python
def single_frame(USER_INPUT_FRAME=False, USER_INPUT_MARGINS=False, frame_number = None, margins = None):
    if USER_INPUT_FRAME:
        frame_number = int(input("Which frame do you want to check?"))

    if USER_INPUT_MARGINS:
        margins = int(input("What are your margins? If 0, no closeup will be supplied"))

    overlay_image_loc(file_input_part_1, file_output_part_1, frame_number, margins)

# For c, frame number is chosen but no margins
single_frame(True)

# For d, frame number is chosen as well as margins
single_frame(True, True)
````
