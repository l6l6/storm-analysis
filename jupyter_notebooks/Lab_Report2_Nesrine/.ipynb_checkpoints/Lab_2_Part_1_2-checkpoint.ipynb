## Part 1.2 : Single Molecule Localization Microscopy (SMLM) Data



```python
from PIL import Image
import matplotlib
import matplotlib.pyplot as pyplot
import os


file_input_part_1_2 = "input/Z calibration (step 10nm).tif"
file_output_part_1_2_h = "output/Z calibration (step 10nm)_LOC_h.hdf5"
file_output_part_1_2_v = "output/Z calibration (step 10nm)_LOC_v.hdf5"

file_output_part_1_2_xml_h = "output/Z calibration (step 10nm)_LOC_h.xml"
file_output_part_1_2_xml_v = "output/Z calibration (step 10nm)_LOC_v.xml"

img = Image.open(file_input_part_1_2)
```

```python
maximum_number_of_frames = -1 # Read all frames

# background params
background_sigma = 10.0
foreground_sigma = 1.0
drift_correction = 0

pixel_size = 100.0
threshold = 6
sigma = 258.21/(100.0 * 2.355)
iterations = 1
roi_size= 11

# Camera params
camera_gain = 1.0
camera_offset = 100.0

# Peak params
find_max_radius = 10
```

```python
import storm_analysis.sa_library.parameters as parameters

def set_params():
    params = parameters.ParametersDAO()

    params.changeAttr("max_frame", maximum_number_of_frames)
    params.changeAttr("start_frame", maximum_number_of_frames)

    params.changeAttr("background_sigma", background_sigma)

    params.changeAttr("camera_gain", camera_gain)
    params.changeAttr("camera_offset", camera_offset)

    params.changeAttr("find_max_radius", find_max_radius, node_type = "float")

    params.changeAttr("foreground_sigma", foreground_sigma)
    params.changeAttr("iterations", iterations)
    params.changeAttr("model", "2d")
    params.changeAttr("pixel_size", pixel_size)
    params.changeAttr("roi_size", roi_size)

    params.changeAttr("sigma", sigma)
    params.changeAttr("threshold", threshold)

    # Tracking
    params.changeAttr("radius", "0.0")
    params.changeAttr("descriptor", "1")


    # No drift-correction
    params.changeAttr("drift_correction", drift_correction)
    return params
```

```python
import storm_analysis.sa_library.datareader as datareader
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import storm_analysis.sa_library.sa_h5py as saH5Py
import matplotlib.pyplot as pyplot
import matplotlib.patches as patches
import numpy

def overlay_image_loc(orig_name, locs_name, frame_number, axis, margins=None, sx = 9, sy = 9):
    """
    Create an image of a frame with the localizations overlaid.

    orig_name - The name of the original file.
    locs_name - The name of the localizations HDF5 file.
    frame_number - Which frame to examine.
    sx - Figure x size in inches.
    sy - Figure y size in inches.
    """
    frame = datareader.inferReader(orig_name).loadAFrame(frame_number).astype(numpy.float64)
    with saH5Py.SAH5Py(locs_name) as h5:
        locs = h5.getLocalizationsInFrame(frame_number)

    frame = frame - numpy.min(frame)
    frame = frame/numpy.max(frame)

    fig = pyplot.figure(figsize = (sx, sy))
    ax = fig.add_subplot(1,1,1)
    fig.subplots_adjust(left=0.2, bottom=0.2)
    for i in range(locs["x"].size):
        width = 10
        height = 10
        if "xsigma" in locs:
            width = height = 5.0*locs["xsigma"][i]
        if "ysigma" in locs:
            height = 5.0*locs["ysigma"][i]
        ellipse = patches.Ellipse((locs["x"][i], locs["y"][i]), width, height, facecolor='none', edgecolor='r', linewidth = 1)
        ax.add_artist(ellipse)

    ax.set_title("Overlay Image on frame {} - axis {}".format(frame_number, axis))
    ax.imshow(frame, interpolation = 'nearest', cmap = "gray")
    ax.scatter(locs["x"], locs["y"], cmap="gray")

    if margins:
        axins = zoomed_inset_axes(ax, 2, loc='lower left', bbox_to_anchor=(0,0), borderpad=3)
        mark_inset(ax, axins, loc1=2, loc2=4)
        axins.set_xlim(margins, margins+20)
        axins.set_ylim(margins, margins+20)
        axins.imshow(frame, cmap = "gray")
        axins.scatter(locs["x"], locs["y"], cmap="gray")
        for i in range(locs["x"].size):
            width = 10
            height = 10
            if "xsigma" in locs:
                width = height = 5.0*locs["xsigma"][i]
            if "ysigma" in locs:
                height = 5.0*locs["ysigma"][i]
            ellipse = patches.Ellipse((locs["x"][i], locs["y"][i]), width, height, facecolor='none', edgecolor='r', linewidth = 1)
            axins.add_artist(ellipse)
    pyplot.show()
```

### a. Sigma horizontal analysis

```python
import storm_analysis.jupyter_examples.overlay_image as overlay_image
import storm_analysis.daostorm_3d.mufit_analysis as mfit

params_x = set_params()
sigma_horizontal = sigma * 0.1

params_x.changeAttr("sigma_range", [sigma_horizontal, sigma])
params_x.toXMLFile(file_output_part_1_2_xml_h, pretty = True)

if os.path.exists(file_output_part_1_2_h):
    os.remove(file_output_part_1_2_h)

mfit.analyze(file_input_part_1_2, file_output_part_1_2_h, file_output_part_1_2_xml_h)
```

```python
frame = 70
margins = 100

overlay_image_loc(file_input_part_1_2, file_output_part_1_2_h, frame, "vertical - horizontal {}".format(round(sigma_horizontal)), margins)
```

### b. Sigma vertical analysis

```python
import storm_analysis.jupyter_examples.overlay_image as overlay_image
import storm_analysis.daostorm_3d.mufit_analysis as mfit

params_y = set_params()
sigma_vertical = sigma * 10

params_y.changeAttr("sigma_range", [sigma, sigma_vertical])
params_y.toXMLFile(file_output_part_1_2_xml_v, pretty = True)

if os.path.exists(file_output_part_1_2_v):
    os.remove(file_output_part_1_2_v)


mfit.analyze(file_input_part_1_2, file_output_part_1_2_v, file_output_part_1_2_xml_v)
```

```python
frame = 70
origins = 100

overlay_image_loc(file_input_part_1_2, file_output_part_1_2_h, frame, "vertical - sigma_vertical {}".format(round(sigma_vertical)), margins)
```

#### c. 3D Position


With the analysis performed in (a) and (b), we assumed the gaussian had independent width in X,Y so we can plot the Wx/Wy versus Z whilst generating the z-offsets of each frame. Then we should apply calibration curves on Z axis.

#### d. Extract Z positions (work in progress)

```python
import numpy

# In this simulation the z range went from -0.6 microns to 0.6 microns in 10nm steps.
z_range = -0.6
z_offsets = numpy.arange(-z_range, z_range + 0.001, 0.01)
valid = numpy.ones(z_offsets.size)

# Limit the z range to +- 0.4um.
mask = (numpy.abs(z_offsets) > 0.4)
valid[mask] = 0.0

numpy.savetxt("z_offsets.txt", numpy.transpose(numpy.vstack((valid, z_offsets))))
```
